```python
import pandas as pd
```

```python
with engine.begin() as conn: # Done with echo=False
    result = conn.execute(text(f'''
                               SELECT * FROM weather_data_hour_seperated; 
                                '''))
    data = result.all()

### Let's create a dataframe out of that
weather_data_hour_seperated = pd.DataFrame(data)#, columns=['...', '...', '...', '...', '...', '...', '...', '...', '...', '...', '...', '...', '...', '...', '...', '...', '...', '...', '...', '...', '...', '...', '...', '...','...']) 
weather_data_hour_seperated
```

```python
nyc_taxi_data_yearly=pd.read_csv('data/nyc_taxi_data_yearly.csv')
```

```python
nyc_taxi_data_yearly
```

```python
# Step 1: Convert 'pu_time' (string) to datetime and extract the hour
nyc_taxi_data_yearly['pu_time'] = pd.to_datetime(nyc_taxi_data_yearly['pu_time'], format='%H:%M:%S')
nyc_taxi_data_yearly['pu_hour'] = nyc_taxi_data_yearly['pu_time'].dt.hour

# Step 2: Ensure 'hour' in weather data is numeric (if it's not already)
weather_data_hour_seperated['hour'] = pd.to_numeric(weather_data_hour_seperated['hour'], errors='coerce')

# Step 3: Convert 'pu_date' to datetime (if it's not already)
nyc_taxi_data_yearly['pu_date'] = pd.to_datetime(nyc_taxi_data_yearly['pu_date'])

# Step 4: Ensure 'date' in weather data is datetime
weather_data_hour_seperated['date'] = pd.to_datetime(weather_data_hour_seperated['date'])

# Step 5: Merge on both 'pu_date' with 'date' and 'pu_hour' with 'hour'
all_tables = pd.merge(
    nyc_taxi_data_yearly,
    weather_data_hour_seperated,
    left_on=['pu_date', 'pu_hour'],  # Merge based on both date and hour
    right_on=['date', 'hour'],      # Merge on date and hour from the weather data
    how='left'                      # Left join
)

```

```python
weather_data_hour_seperated['time'] = weather_data_hour_seperated['time'].dt.tz_localize(None)
weather_data_hour_seperated['time']
```

```python
nyc_taxi_data_yearly['pu_hour'] = nyc_taxi_data_yearly['pu_time'].dt.hour
nyc_taxi_data_yearly['pu_hour']
```

```python
all_tables
```

```python
# Specify the path where you want to save the CSV file
file_path = 'data/nyc_taxi_weather_data.csv'

# Save the DataFrame to CSV
all_tables.to_csv(file_path, index=False)

print(f"DataFrame saved as CSV at {file_path}")
```

```python

url = f'postgresql://{pg_user}:{pg_pass}@{pg_host}:{pg_port}/{pg_db}'
engine = create_engine(url, echo=False)

# Specify the name of the table with the special schema
schema_name = 'for_o_for_wizards'
table_name = 'nyc_taxi_weather_data'

# Attempt to upload the DataFrame to PostgreSQL (create table or append)
try:
    all_tables.to_sql(table_name, engine, schema_name, if_exists='replace', index=False, chunksize=5000)
    print(f"Data uploaded successfully to {table_name}.")
except Exception as e:
    print(f"An error occurred: {e}")
```
